```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
Aotizhongxin = pd.read_csv("data\Air-quality-dataset\PRSA_Data_20130301-20170228\PRSA_Data_Aotizhongxin_20130301-20170228.csv")
Changping = pd.read_csv("data\Air-quality-dataset\PRSA_Data_20130301-20170228\PRSA_Data_Changping_20130301-20170228.csv")
Dingling = pd.read_csv("data\Air-quality-dataset\PRSA_Data_20130301-20170228\PRSA_Data_Dingling_20130301-20170228.csv")
Dongsi = pd.read_csv("data\Air-quality-dataset\PRSA_Data_20130301-20170228\PRSA_Data_Dongsi_20130301-20170228.csv")
Guanyuan = pd.read_csv("data\Air-quality-dataset\PRSA_Data_20130301-20170228\PRSA_Data_Guanyuan_20130301-20170228.csv")
Gucheng = pd.read_csv("data\Air-quality-dataset\PRSA_Data_20130301-20170228\PRSA_Data_Gucheng_20130301-20170228.csv")
Huairou = pd.read_csv("data\Air-quality-dataset\PRSA_Data_20130301-20170228\PRSA_Data_Huairou_20130301-20170228.csv")
Nongzhanguan = pd.read_csv("data\Air-quality-dataset\PRSA_Data_20130301-20170228\PRSA_Data_Nongzhanguan_20130301-20170228.csv")
Shunyi = pd.read_csv("data\Air-quality-dataset\PRSA_Data_20130301-20170228\PRSA_Data_Shunyi_20130301-20170228.csv")
Tiantan = pd.read_csv("data\Air-quality-dataset\PRSA_Data_20130301-20170228\PRSA_Data_Tiantan_20130301-20170228.csv")
Wanliu = pd.read_csv("data\Air-quality-dataset\PRSA_Data_20130301-20170228\PRSA_Data_Wanliu_20130301-20170228.csv")
Wanshouxigong = pd.read_csv("data\Air-quality-dataset\PRSA_Data_20130301-20170228\PRSA_Data_Wanshouxigong_20130301-20170228.csv")

data = {
    "AO": Aotizhongxin,
    "CH": Changping,
    "DI": Dingling,
    "DO": Dongsi,
    "GY": Guanyuan,
    "GC": Gucheng,
    "HU": Huairou,
    "NZ": Nongzhanguan,
    "SH": Shunyi,
    "TI": Tiantan,
    "WL": Wanliu,
    "WS": Wanshouxigong
}
```

```python
df = [
    data.get("AO"), data.get("CH"), data.get("DI"), 
    data.get("DO"), data.get("GY"), data.get("GC"), 
    data.get("HU"), data.get("NZ"), data.get("SH"), 
    data.get("TI"), data.get("WL"), data.get("WS")
]

df = pd.concat(df)
```

- Data terdiri dari 420768 baris dan 18 kolom

### Assessing Data

```python
df.duplicated().sum()
```

**Insight:**
- Tidak terdapat nilai duplikat
- Terdapat beberapa kolom yang memiliki nilai null

### Cleaning Data

```python
df.isna().sum()
```

```python
df = df.dropna(subset=["TEMP", "PRES", "DEWP", "RAIN", "WSPM", "wd"])
```

```python
df["PM2.5"] = df["PM2.5"].interpolate(method='linear', limit_direction='forward')
df["PM10"] = df["PM10"].interpolate(method='linear', limit_direction='forward')
df["SO2"] = df["SO2"].interpolate(method='linear', limit_direction='forward')
df["NO2"] = df["NO2"].interpolate(method='linear', limit_direction='forward')
df["CO"] = df["CO"].interpolate(method='linear', limit_direction='forward')
df["O3"] = df["O3"].interpolate(method='linear', limit_direction='forward')
```

**Insight:**
- Kolom yang memiliki jumlah nilai null minor (<2000 baris) akan dihapus
- Kolom yang memiliki jumlah nilai null mayor (>2000 baris) akan diisi dengan menggunakan metode interpolasi linear agar data tidak kehilangan informasi

## Exploratory Data Analysis (EDA)

### Explore ...

```python
plt.figure(figsize=(15, 8))
plt.title("Korelasi Tiap Kolom pada Dataset")
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".0%")
plt.show()
```

```python
numeric_columns = df.drop(columns=["No", "year", "month", "day", "hour"])

plt.figure(figsize=(15, 8))

numeric_columns = numeric_columns.select_dtypes(include=['number']).columns
num_columns = len(numeric_columns)
num_rows = (num_columns + 1) // 2

for i, column in enumerate(numeric_columns): 
    plt.subplot(num_rows, 2, i + 1)  
    sns.histplot(df[column], kde=True) 
    plt.title(column) 
    plt.xlabel('Value')  
    plt.ylabel('Frequency') 

plt.tight_layout() 
plt.show() 
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana tren level polusi udara di tiap wilayah berubah?

```python
sns.set_theme(style="darkgrid", palette="pastel")
plt.figure(figsize=(20, 10))
plt.title("Time Series Plot of PM2.5 for Each Station")
sns.lineplot(x="year", y="PM2.5", hue="station", data=df, marker='o', errorbar=None)
plt.xlabel("Year")
plt.ylabel("PM2.5")
plt.xticks(df['year'].unique())
plt.yticks(range(60, 120, 20))
plt.legend(loc="upper left", ncol=3)
plt.show()
```

### Pertanyaan 2: Bagaimana perbandingan level polusi udara di tiap wilayah?

```python
average_pm25 = df.groupby('station')['PM2.5'].mean().reset_index()
plt.figure(figsize=(15, 8))
sns.barplot(x='station', y='PM2.5', data=average_pm25, palette='pastel')
plt.yticks(range(0, 100, 5))
plt.title("Average PM2.5 Levels by Station")
plt.xlabel("Station")
plt.ylabel("Average PM2.5 Levels")
plt.show()
```

**Insight:**
- Terdapat peningkatan signifikan pada tahun 2016-2017
- Penurunan level polusi udara terjadi dari tahun 2015 hingga bertemu titik terendahnya pada tahun 2016
- Wilayah yang memiliki level polusi udara tertinggi adalah wilayah Dongsi, diikuti oleh wilayah Nongzhanguan dan Wanshouxigong

```python
beijing = gpd.read_file("https://raw.githubusercontent.com/d3cn/data/refs/heads/master/json/geo/china/province-city/beijing.geojson")
beijing = beijing.to_crs("EPSG:4326")

fig, ax = plt.subplots(figsize=(15, 8))
beijing.plot(ax=ax, color="lightgrey")
ax.set_title("Average PM2.5 Levels by Station in Beijing")
ax.set_axis_off()

scatter = ax.scatter(
    x=coordinates_and_pm25["longitude"],
    y=coordinates_and_pm25["latitude"],
    s=coordinates_and_pm25["PM2.5"] * 10,
    c=coordinates_and_pm25["PM2.5"],
    cmap="coolwarm",
    alpha=0.7,
    edgecolors="w",
    linewidth=1
)

for i, txt in enumerate(coordinates_and_pm25["station"]):
    ax.annotate(txt, (coordinates_and_pm25["longitude"][i], coordinates_and_pm25["latitude"][i]), fontsize=10)

cbar = plt.colorbar(scatter)
cbar.set_label("Average PM2.5 Levels")
plt.show()
```
